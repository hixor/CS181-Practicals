```python
def getMolsFromSmiles(smileset,filename):
    mol_set = []
    i=0
    for smile in smileset:
        mol = Chem.MolFromSmiles(smile)
        mol_set.append(mol)
        i=i+1
        if i%1000==0:
            print "!000 done."
    np.savetxt(filename,mol_set,delimiter=',')
    return mol_set
```

```python
def getRDKFingerprints(transform_filename, data):
    tf = open(transform_filename, 'wb')

    # Produce a CSV File:
    transform_csv = csv.writer(tf, delimiter=',', quotechar='"')

    prog = 0
    for index, datum in enumerate(data):
        prog += 1
        if prog % 10000 == 0:
            print prog
        transform_csv.writerow(np.hstack((
                                datum,
                                RDKFingerprint(datum,fpSize=512),
                                ))
                               )
    print "done"
```

```python
def getMACCS(transform_filename, data):
    tf = open(transform_filename, 'wb')

    # Produce a CSV File:
    transform_csv = csv.writer(tf, delimiter=',', quotechar='"')

    prog = 0
    for index, datum in enumerate(data):
        prog += 1
        if prog % 10000 == 0:
            print prog
        transform_csv.writerow(np.hstack((
                                datum,
                                MACCSkeys.GenMACCSKeys(datum),
                                ))
                               )
    print "done"
```

```python
def getMorgan(transform_filename, data):
    tf = open(transform_filename, 'wb')

    # Produce a CSV File:
    transform_csv = csv.writer(tf, delimiter=',', quotechar='"')

    prog = 0
    for index, datum in enumerate(data):
        prog += 1
        if prog % 10000 == 0:
            print prog
        transform_csv.writerow(np.hstack((
                                datum,
                                AllChem.GetMorganFingerprint(datum,2,invariants=[1]*datum.GetNumAtoms()),
                                ))
                               )
    print "done"
```

```python
# Write a prediction file.
def write_prediction_file(pred_filename, data, clf):
    pred_fh = open(pred_filename, 'wb')

    # Produce a CSV file.
    pred_csv = csv.writer(pred_fh, delimiter=',', quotechar='"')

    # Write the header row.
    pred_csv.writerow(['Id', 'Prediction'])

    prog = 0
    for index, datum in enumerate(data):
        prog += 1
        if prog % 5 == 0:
            print prog
        pred_csv.writerow([index, clf.predict(datum)[0]])
    print "done"
```
